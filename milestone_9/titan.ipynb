```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
# check survival status by gender
sns.displot(titanic, x = 'survived', hue = "sex")
```

```python
# check survival status by pclass
sns.displot(titanic, x = 'survived', hue = "pclass")
```

```python
# display count of passengers by gender
sns.displot(titanic, x = 'sex')
```

```python
# display count of passengers by pclass
sns.displot(titanic, x = 'pclass')
```

```python
# conclusion
# majority of passengers were men but in percentage terms more women survived
# majority of passengers were from 3 class but in percentage terms more survived from 1 class
```
