```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# check at what time the amount of the check is greater
sns.displot(tips, x = 'total_bill', hue = 'time')
```

```python
# checking for sex who leaves a bigger tip
sns.violinplot(tips, x='sex', y='tip')
```

```python
# check on which day of the week the tip is bigger
sns.displot(tips, x = 'tip', hue = "day", kind = "kde")
```

```python
# check at what time of day the tip is bigger
sns.displot(tips, x = 'tip', hue = "time", kind = "kde")
```

```python
# check which gender has a bigger check
sns.displot(tips, x = 'total_bill', hue = "sex", kind = "kde")
```

```python
# conclusion
# tips are higher in the evenings and on weekends
# men have higher number of checks and total bill
# at a time when the bigger the check, the bigger the tip
#
```
